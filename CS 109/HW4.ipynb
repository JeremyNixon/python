#### Problem 2(a)

Assume you have $M$ polls with sample sizes $n_1, \dots, n_M$. If the polls are independent, what is the average of the variances of each poll if the true proportion is $p$?

** Your answer here: **

With independent polls, the variance is p(1−p)/N. The average of the variances of each poll will be 1/M times the sum from i = 1 to M of( (pi(1-pi)/N).

$$\mbox{Avg Var}(\bar{X})= \frac{1}{M} \sum_{i \in n}\frac{p(1-p)}{N} = \frac{p(1-p)}{M} \sum_{i \in n}\frac{1}{N} $$

We have a harmonic series that diverges as n increses. 

```python
polls = dict((election.split("/")[-1][:-4], build_frame(election)) for election in election_url)
variance1 = []
for keys in polls:
    data2 = polls[keys]
    number_of_observations = data2.columns[4]
    N = data2[number_of_observations].mean()
    candidate1 = data2.columns[7]
    candidate2 = data2.columns[8]
    p1 = (data2[candidate1]).mean()/100
    p2 = (data2[candidate2]).mean()/100
    var1 = p1 * (1-p1) / N 
    var2 = p2 * (1-p2) / N 
    variance1.append(var1)
variance1
```

#### Problem 3(a)

First, we define a Bayesian model for each race. The prior for the difference $d$ between republicans and democtrats will be $N(\mu,\tau^2)$. Say before seeing poll data you have idea who is going to win, what should $\mu$ be? How about $\tau$, should it be large or small? 

** Your answer here: **
$\mu$ Should be your prior, the original probability that you give to the chances of your hunch winning. Even if we have no poll data about $\mu$, we may think that incumbents have some different prior or have some other useful information. Without concrete prior information, we set $\mu$ equal to 0. $\tau$ is the variability of each poll, here the variance of the difference between the percentage of Republican and Democrat votes. Large $\tau$ means an increased possibility of big differences between results. In general results go close to prediction and we have little reason to think that there will be big swings in election results, so we should assume small $\tau$ for this distribution.
