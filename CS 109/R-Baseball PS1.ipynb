```python
%matplotlib inline

# Rules for learning - no copy/paste, no same variable names

import requests as rq
import StringIO as st
import zipfile as zp
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
def manipulate_zip_file(name_of_zip_file):
    request = rq.get(name_of_zip_file).content
    convert_to_string = st.StringIO(request)
    zipfile = zp.ZipFile(convert_to_string,'r')
    return zipfile
```

```python
address = 'http://seanlahman.com/files/database/lahman-csv_2014-02-14.zip'
zipfile = manipulate_zip_file(address)
```
