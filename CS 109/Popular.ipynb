### Obtaining Data from Audio Files

**1. Looking at WAV File Composition**

First, we explore a wav file to see what the data looks like

```python
# Define function to give us the left and right sides of a file and the frequencies 
def analyze_wav(filename):
    """
    Returns two numpy arrays, the left and right sides of the wav, 
    and the frequency as an int
    """
    # Import the file and normalize it
    s, fs, enc = ad.wavread(filename)  # s in range -32768 to +32767
    s = np.array(s)
    s = s / abs(s).max() 
    
    # Divide into two sides
    left = s[:,0]
    right = s[:,1]
    
    return left, right, fs
```

**2. Building a Dataset of All Relevant Audio Files**

Moving forward, we can't keep both the right and left sides. Therefore, we have to store all the right sides of each wave in a large matrix, called S. We also save the relative frequencies in a large matrix called F. 

```python
def create_S_matrix(folder):
    """
    Returns two numpy arrays, S and F, the array of actual data points and the array of frequencies
    """
    S = []
    F = []
    names = []

    for direc in os.listdir(folder):
         if not os.path.isfile(direc):
            newfolder = folder + "/" + direc
            for fn in os.listdir(newfolder):
                if not os.path.isfile(fn):
                    filename = newfolder + "/" + fn
                    left, right, fs = analyze_wav(filename)
                    S.append(right)
                    F.append(fs) 
                    names.append(fn[:-4])
    S = np.array(S)
    
    return S, F, names
```

**4. Using the FFT **

Now it's time to manipulate our wave functions. To do this, we have to run an "fft". First, we use the Discrete Fourier Transform to extract the frequencies of our signal.

```python
"""
Find 2^n that is equal to or greater than.
"""
def nextpow2(i):
    n = 1
    while n < i: n *= 2
    return n
```

```python
def fft(S, S_shifted, length):
    """
    Calculates fft for S and S_shifted
    """
    X = []
    f = []
    X_shifted = []

    m = S.shape[0]
    NFFT = nextpow2(length)
    t = np.linspace(0,1, NFFT/2+1)

    for i in range(m):
        X.append(scipy.fft(S[i]))
        f.append(scipy.fftpack.fftfreq(S[i].size, t[1]-t[0]))
        X_shifted.append(abs(scipy.fft(S_shifted[i]))/length)
    
    return X, f, X_shifted
```

**5. Compression of Matrices**

Moving forward, we're going to work with something a bit simpler. 220500 x 220500 matrices are just too big to work with! We need to compress our music down. Thus, we compress our files to just include the highest 5000 frequencies, replacing the other frequencies with zeros.

```python
def compress(X):
    """
    Compresses the output of an fft down to the highest 5000 frequencies
    """
    Comp_X = np.array(X)
    m, n = Comp_X.shape
    for i in range(m):
        B = np.array(abs(Comp_X[i,:]))
        B.sort()
        np.putmask(Comp_X[i,:], abs(Comp_X[i,:]) < B[n-5001], 0)
    
    return Comp_X
```

**6. Finding an eigenset**

Ok, we've compressed our vectors. Let's chop off all the zeros (we'll explain why we can do this later) and find the covariance matrix. For our purposes, we compressed from 220,500 to 5,000, and then cut off the zeros after 10,000.

```python
def covariance(Comp_X):
    """
    Finds the covariance matrix of a compressed matrix
    """
    New_X = Comp_X[:,0:10000]
    covariance = np.transpose(np.matrix(New_X))*np.matrix(New_X)
    return covariance
```

After creating the covariance function, we now calculate the covariance matrix for both popular songs and unpopular songs.

```python
covariance_unpop = covariance(Comp_Unpop)
```

** 8. Extracting the Important Datapoints**

Now that we have the ability to extract frequency values for a given song, we extract the datapoints we will use for our classification:

```python
# Function to calculate datapoints for a matrix of ffts 
def datapoints(Comp_X, names):
    """
    Calculates the mean, median, standard deviation and maximum frequency of array of fft values for songs.
    The array must be m x n sized, where m is the number of songs and n is the length of the fft for one song.
    Returns an array of mean, median, standard deviation and maximum frequency values.
    """
    data = []
    n = Comp_X.shape[0]
    for i in range(n):
        p = np.array(np.nonzero(abs(Comp_X[i])))
        mean = np.mean(p)
        median = np.median(p)
        stddev = np.std(p)
        maxf = p.max()
        cov = np.array(np.matrix(p)*np.transpose(np.matrix(p)))
        data.append([mean, median, stddev, maxf, cov[0][0], names[i]])
    data = pd.DataFrame(data, columns = ["Mean", "Median", "StdDev", "Max", "Covariance", "Song Name"])
    return data
```

However, in working with this data, we need to work with numbers, not strings. Therefore, we convert all the strings in our dataframe into integers:

```python
def turn_into_numbers(DataFrame):
    """
    Takes a dataframe with the columns listed above and converts the entries into numbers
    """
    Artist = []
    SongName = []
    Genre = []
    Key = []

    for value in DataFrame['Artist']:
        Artist.append(sum([ord(char) for char in value]))
    for value in DataFrame['Song Name']:
        SongName.append(sum([ord(char) for char in value]))
    for value in DataFrame['Genre']:
        if value == 'Country':
            Genre.append(0)
        elif value == 'Rock':
            Genre.append(1)
        elif value == 'Hip Hop':
            Genre.append(2)
        elif value == 'Pop':
            Genre.append(3)
        elif value == 'EDM':
            Genre.append(4)
        else:
            Genre.append(5)
    for value in DataFrame['Key']:
        Key.append(sum([ord(char) for char in value]))
    EditedDataFrame = pd.DataFrame([list(DataFrame['Year']), Artist, SongName, Genre, list(DataFrame['Tempo']), Key, 
                                           list(DataFrame['Mean']), list(DataFrame['Median']), list(DataFrame['StdDev']),
                                           list(DataFrame['Max']), list(DataFrame['Covariance'])]).T
    EditedDataFrame.columns = [DataFrame.columns]
    return EditedDataFrame
```

```python
"""
cutoff_predict(clf, X, cutoff)

Inputs:
clf: a **trained** classifier object
X: a 2D numpy array of features
cutoff: a float giving the cutoff value used to convert
        predicted probabilities into a 0/1 prediction.

Output:
a numpy array of 0/1 predictions.
"""
def cutoff_predict(clf, X, cutoff):
    Predicted_Probabilities = clf.predict_proba(X)
    predictions = []
    for row in Predicted_Probabilities:
        if row[1] > cutoff:
            predictions.append(1)
        else:
            predictions.append(0)  
    predictions = np.array(predictions)
    return predictions

def custom_f1(cutoff):
    def f1_cutoff(clf, X, y):
        ypred = cutoff_predict(clf, X, cutoff)
        return sklearn.metrics.f1_score(y, ypred)
    return f1_cutoff
```

```python
# Define a function that will be able to plot the decision surface
# Note: Code in this section borrows heavily fro Homework 5
def plot_decision_surface(plot_step, clf, X_train, Y_train):
    
    if X_train.shape[1] != 2:
        raise ValueError("X_train should have exactly 2 columnns!")
    
    x_min, x_max = X_train[:, 0].min() - plot_step, X_train[:, 0].max() + plot_step
    y_min, y_max = X_train[:, 1].min() - plot_step, X_train[:, 1].max() + plot_step
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    clf.fit(X_train,Y_train)
    if hasattr(clf, 'predict_proba'):
        Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
    else:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])    
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Reds)
    plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=plt.cm.Paired)
    plt.show()
```

## Final Data Analysis

The takeaways for the project can be grouped in the aforementioned categories.

**Historical Analysis** 

From our historical analysis, we inferred that the spread of mean frequencies among the most popular songs is roughly the same every year. However, we discovered that most years are severely skewed in one direction. This demonstrates that while in a given set of popular songs you are likely to find hits at both edges of the distribution, the general means may tend towards one edge. One possible explanation for this is that tastes vary from year to year and that is represented in the frequencies of the song. Another possibility is that because a given album can have many similar top hits, that there is unaccounted-for correlation about the top 25 songs for a given year—i.e. they are from the same artists. As a consequence, these songs are likely to have similar frequency statistics and may explain the change from one year to another.

**Classical/Popular Classifer**

From our classical/popular classifier, we learned that it is very easy to express the differences between classical and popular music. Our model was able to very quickly identify the differences between the two and quickly approached an F1 score of 1 with no variation. For the F1 score, the harmonic mean of recall and precision, to be 1, we can infer that both recall and precision are 1—indicating that we correctly guess every positive without have erroneous guesses. Our model is thus, essentially, perfect. A possible reason for this is that the two datasets were very different because classical and popular music comes from very different instruments. As a result, a lot of the frequency metrics that we end up looking at will be significantly different between the two and, consequently, it will be significantly easier to differentiate between the two.

**Genre Classifier**

The genre classifier that we built was not great, but we were able to see that the artist has no impact. This was a surprising result, since we would have assumed that the artist would be a primary determinant of the genre since most artists tend to produce one genre only (Taylor Swift, whose music maintained popularity even as it shifted genres from country to pop, is one exception to this rule). Overall, the F1 scores that we generated were not high enough to have a good degree of confidence in our model.
 
**Popular/Unpopular Classifier**

Our popular/unpopular classifier did not have the same success as the first one. By contrast, we found that using underlying frequency statistics did not help predict which songs would become popular for a given artist. None of the frequency characteristics that we considered—covariance, max, mean, median, standard deviation—were judged to be particularly predictive. This could be a product of popularity coming less from musical differences and more because of other factors (lyrics, momentum factors, etc.). On the other hand, it is possible that we simply did not have enough data and had we included more unpopular songs we would have been able to draw a difference between the unpopular and popular sets. Another possibility is that popular musicians tend to necessarily have less variety in their music than one-hit wonders, whose single hit may differ significantly from its typical work.
