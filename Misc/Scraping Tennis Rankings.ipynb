```python
# Scraping Historical Tennis Rankings
# David Freed, Barrett Hansen, Jayant Sani
# November 29, 2014

def getATPtourDates():
    '''
    Inputs: None
    Outputs: List of dates for which the ATPtour website has rankings data
    '''
    
    # Scrape website to grab all possible dates
    base = 'http://www.atpworldtour.com/Rankings/Singles.aspx?d='
    dates = urllib2.urlopen(base+str('&r=1&c=#')).read()

    # Convert to Beautiful Soup
    datesoup = BeautifulSoup(dates)

    # Grab all the possible dates as a list
    datelist = datesoup.select('form')[1].select('option')
    dates = [item.contents[0] for item in datelist[0:-1]]
    return(dates)
```

```python
def ATPrankingsScraper(dates):
    '''
    Inputs: List of dates over which to scrape Rankings data
    Outputs: Pandas df containing: 
        - Player Name
        - Player Ranking
        - Change in Ranking since last Rankings
        - URL for player information on ATPtour.com
    '''
    
    base = 'http://www.atpworldtour.com/Rankings/Singles.aspx?d='

    
    # Create empty dataframe with which to concatenate
    y = pd.DataFrame(columns=['Rank', 'First Name', 'Last Name', 'Points', 
                              'Change', 'Date', 'Player URL'])

    # Loop over the trial dataset to see if code works
    for item in dates:

        # Create the proper URL to scrape
        full = base + item + str('&r=1&c=#')

        # Use LXML to grab the dynamically-generated CSS from the website
        x = requests.get(full)
        tree = lxml.html.fromstring(x.text)

        # Select the table from the new CSS
        sel = CSSSelector('table')
        results = sel(tree)

        # For simple scraping, just use the Beautiful Soup package
        soup123 = BeautifulSoup(lxml.html.tostring(results[0]))

        # Create arrays in which to load in all the data
        rank, firsts, lasts, points, change, links = [], [], [], [], [], []

        # Loop through the table and input all of the relevant information
        for tr in soup123.find_all('tr'):
            if len(tr.find_all('a')) != 0:
                rank.append(int(tr.find_all('span')[0].contents[0]))
                points.append(int(tr.find_all('a')[1].contents[0].replace(',', '')))
                change.append(int(tr.find_all('td')[2].contents[0]))

                # Do slight string parsing for player name
                name = tr.find_all('a')[0].contents[0].replace(u'\xa0',u' ')
                link = tr.find_all('a')[0].get('href')
                first = name.split(",")[1][1:] 
                last =  name.split(",")[0]
                firsts.append(first)
                lasts.append(last)
                links.append(link)

        # Turn the arrays into a dataframe, renaming the columns and inserting a date
        ranking = pd.DataFrame(data = [rank, firsts, lasts, points, change]).T
        ranking.columns = ['Rank', 'First Name', 'Last Name', 'Points', 'Change']
        ranking['Date'] = item
        ranking['Player URL'] = ['http://www.atpworldtour.com'+link for link in links]

        #print item

        # Concatenate with previous DataFrames
        y = pd.concat([y,ranking], ignore_index=True)
    
    # Re-order all the columns for aesthetic purposes
    y = y[['Date', 'Rank', 'First Name', 'Last Name', 'Points','Change', 'Player URL']]
    
    # Create dictionary that contains
    
    # Turn results into a .csv
    y.to_csv("ranking_history.csv")
    return(y)
```

```python
def ATPplayerInfo(url):
    '''
    Inputs: url of ATPtour player bio page
    Outputs: Pandas df containing variety of player bio information
    '''
    #print(url)
    
    # Read website and convert to beautiful soup
    website = urllib2.urlopen(url)
    html = website.read()
    soup = BeautifulSoup(html)
    
    # Scrape the headers and contents of the bio page
    headers = [str(el.contents[0].contents[0])[0:-1] for el in 
           soup.find_all("ul", attrs={"id": "playerBioInfoList"})[0].find_all('li')]
    contents = [str(el.contents[1])[1:] for el in 
            soup.find_all("ul", attrs={"id": "playerBioInfoList"})[0].find_all('li')]
    
    # Convert to Pandas DF and reformat to ints (use cms and kgs for height and weight)
    info_df = pd.DataFrame(contents).T
    info_df.columns = headers
    
    if 'Age' in headers:
        info_df['Birthdate'] = info_df['Age'][0][4:-1]
        info_df['Age'] = int(info_df['Age'][0][0:3])
    
    # Height and Weight with re.compile
    parens = re.compile('\(')
    if 'Height' in headers:
        h_start = parens.search(info_df['Height'][0]).end()
        info_df['Height'] = int(info_df['Height'][0][h_start:h_start+3])
    
    if 'Weight' in headers:
        w_start = parens.search(info_df['Weight'][0]).end()
        info_df['Weight'] = int(info_df['Weight'][0][w_start:-4])
    
    # Add name back in
    info_header = soup.find_all(attrs={"id": "playerBioInfoCardHeader"})[0]
    info_df['Player Name'] = str(info_header.find_all('h1')[0].contents[0])
    
    #print(info_df['Player Name'])
    return(info_df)
```

```python
def ATPplayersInfo(players):
    '''
    Inputs: list of ATPtour.com player bio links
    Outpus: Pandas df of player information
    *Note: Height and weight figures will be ones most recently found on site
    '''
    
    # Intialize dataframe
    headers = ['Player Name', 'Age', 'Birthdate', 'Birthplace', 'Residence', 
                       'Height', 'Weight', 'Plays', 'Turned Pro', 'Coach']
    y = pd.DataFrame(columns = headers)
    
    # Eliminate non-unique entries in list
    players = list(set(players))
    
    # Loop over player urls in list
    for player in players:
        #print(player)
        tmp_df = ATPplayerInfo(player)
        y = pd.concat([y, tmp_df])
    
    y = y[['Player Name', 'Age', 'Birthdate', 'Birthplace', 'Residence', 
                       'Height', 'Weight', 'Plays', 'Turned Pro', 'Coach']]
    
    y.reindex()
    
    return(y)
```

```python
def ATPrankingsAndPlayers(dates, playerURLsVisited):
    '''
    Inputs: List of dates and List of URLs previously visited (can be empty)
    Outputs: Dictionary containing:
        - Rankings: pandas df of ATP Top 100 rankings back to 1973
        - Players: pandas df of all players to have made the ATP Top 100 since 1973
    '''
    
    # Initialize the dataframes
    #rankings = ATPrankingsScraper(getATPtourDates()[0:1])
    headers1 = ['Player Name', 'Age', 'Birthdate', 'Birthplace', 'Residence', 
                       'Height', 'Weight', 'Plays', 'Turned Pro', 'Coach']
    headers2 = ['Rank', 'First Name', 'Last Name', 'Points', 
                              'Change', 'Date', 'Player URL']
    players = pd.DataFrame(columns=headers1)
    rankings = pd.DataFrame(columns=headers2)
    
    # Get all the dates, and determine number of iterations needed to run
    #rankings_obs = len(dates)
    #iterations = int(np.ceil(float(rankings_obs)/float(400)))
    
    # Initialize empty list that tracks players visited
    #playerURLsVisited = []
    
    # Iterate through so request doesn't time out
    rankings = ATPrankingsScraper(dates)
    playerURLs = list(set(rankings['Player URL']))
    playerURLs = [el for el in playerURLs if el not in playerURLsVisited]
    players = ATPplayersInfo(playerURLs)

    # Create dictionary that stores two dfs
    d = {}
    d['Players'] = players
    d['Rankings'] = rankings
    d['URLsVisited'] = playerURLs
    
    return(d)
```
