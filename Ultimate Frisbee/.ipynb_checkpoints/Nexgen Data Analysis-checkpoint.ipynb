** Question: **
Let's figure out what positions on the field are associated with important events (turns, mostly) and see what events lead to those positions.

```python
plays['x'] = plays['xPosition']
plays['y'] = plays['yPosition'] * (4/11.0)
plays.head()
```

```python
def binning(data,bins):
    arr = []
    size = data.max()
    for n in range(0,len(data)):
        arr.append(int(math.floor(data[n]/(size/bins))))
    return arr

binWidth = 1
xBins = 110/binWidth
yBins = 40/binWidth
plays['xBin'] = binning(plays['x'], xBins)
plays['yBin'] = binning(plays['y'], yBins)

array = []
for i in range(0,xBins + 1):
    temp = []
    for j in range(0,yBins + 1):
        temp.append(0)
    array.append(temp)

for n in range(0, len(plays)):
    if plays['type'][n] == 'POINT':
        array[plays['xBin'][n]][plays['yBin'][n]] += 1
```
