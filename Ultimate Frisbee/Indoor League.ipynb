```python
file = "/Users/Jeremy/python/Ultimate Frisbee/Indoor.csv"
dataset = pd.read_csv(file)
dataset.head()
```

```python
dataset = dataset[(dataset['Games'] > 0)]
dataset['GPG'] = dataset['Goals']/dataset['Games']
dataset['APG'] = dataset['Assists']/dataset['Games']
dataset['DPG'] = dataset['Ds']/dataset['Games']
dataset['TPG'] = dataset['Turns']/dataset['Games']
dataset['DrPG'] = dataset['Drops']/dataset['Games']
dataset['PBPG'] = dataset['Point Blocks']/dataset['Games']
dataset['Performance'] = (dataset['GPG'] + dataset['APG']) / (dataset['DPG'] + dataset['TPG'])
dataset['ATR'] = (dataset['APG'])/(dataset['TPG']+(3/dataset['Games']))
# Constant 3 to reduce impact of low turn outliers
dataset['GATR'] = (dataset['APG']+dataset['GPG'])/(dataset['TPG']+(3/dataset['Games']))
dataset['GATDR'] = (dataset['APG']+dataset['GPG'])/(dataset['TPG']+(3/dataset['Games'])+dataset['Drops'])
dataset['GADTDR'] = (dataset['APG']+dataset['GPG'])*(dataset['DPG'])/(dataset['TPG']+(3/dataset['Games'])+dataset['Drops'])
dataset['GA'] = (dataset['Goals']+dataset['Assists'])
dataset['TD'] = (dataset['Turns']+dataset['Drops'])
dataset.head()
```

```python
def get_count(item):
    counts = {}
    for x in item:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
def top_counts(item, n=5):
    pairs = [(count, tz) for tz, count in item.items()]
    pairs.sort()
    return pairs[-n:]
```
